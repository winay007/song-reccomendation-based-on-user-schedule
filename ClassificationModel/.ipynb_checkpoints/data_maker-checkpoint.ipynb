```python
import joblib
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split 
```

```python
model = joblib.load("xg_model.pkl")
model1 = joblib.load("model1.pkl")
model2 = joblib.load("model2.pkl")
model3 = joblib.load("model3.pkl")
final_model = joblib.load("final_model.pkl")
```

```python
df = pd.read_csv("data/data.csv")
```

```python
col_features = ["danceability","energy","acousticness","acousticness","instrumentalness","liveness","loudness","speechiness","tempo"]
# X= MinMaxScaler().fit_transform(df[col_features])
X2 = np.array(df[col_features])
X2
```

```python
pred1 = model1.predict(X2)
pred2 = model2.predict(X2)
pred3 = model3.predict(X2)
```

```python
stack_data = np.column_stack((pred1,pred2,pred3))
```

```python
prediction = final_model.predict(stack_data)
```

```python
prediction = prediction.astype(str)
prediction[prediction == "0"] = "Calm"
prediction[prediction == "1"] = "Enegetic"
prediction[prediction == "2"] = "Happy"
prediction[prediction == "3"] = "Sad"
prediction
```

```python
df['mood'] = prediction
```

```python
df
```

```python
df.to_csv("data/classifiedSong.csv",index=False)
```
